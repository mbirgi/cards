```python
import cv2
import matplotlib.pyplot as plt

def grid_display(list_of_images, list_of_titles=[], no_of_columns=2, figsize=(10,10)):

    fig = plt.figure(figsize=figsize)
    column = 0
    for i in range(len(list_of_images)):
        column += 1
        #  check for end of column and create a new figure
        if column == no_of_columns+1:
            fig = plt.figure(figsize=figsize)
            column = 1
        fig.add_subplot(1, no_of_columns, column)
        plt.imshow(list_of_images[i])
        plt.axis('off')
        if len(list_of_titles) >= len(list_of_images):
            plt.title(list_of_titles[i])

img_1 = cv2.imread('./img/80px-Playing_card_club_A.svg.png')
img_2 = cv2.imread('./img/80px-Playing_card_club_2.svg.png')
images = [img_1, img_2]
grid_display(images, figsize=(2,2))
```

```python
import os
import random

path = 'img'
deck = [os.path.join(path, f.name) for f in os.scandir(path) if f.is_file() and f.name != '.DS_Store']
# display(deck)
random.shuffle(deck)
images = [cv2.imread(card) for card in deck]
grid_display(images, no_of_columns=13, figsize=(18,5))
```
